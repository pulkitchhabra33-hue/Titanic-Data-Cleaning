```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()

   
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df

```

```python
df.drop(columns= ['Cabin'], inplace=True)
```

```python
df.columns
```

```python

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace= True)
```

```python
df.drop_duplicates(inplace= True)
```

```python
df['Survived'].value_counts()   
```

```python
print("Survival Counts")
print(df['Survived'].value_counts())
```

```python
print("Gender Counts")
print(df['Sex'].value_counts())
print()
```

```python
print("Survival rate by gender: ")
sv= df.groupby('Sex')['Survived'].mean()
sv
```

```python
%matplotlib inline
plt.hist(sv)
```

```python
class_counts= df.groupby(['Pclass', 'Survived']).size().unstack(fill_value=0)

class_counts.plot(kind= 'bar', stacked= False)
plt.title('Survival Counts by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(['Did not survive(0)', 'Survived(1)'])
plt.show()
```

```python
df['Age']= pd.to_numeric(df['Age'], errors= 'coerce')
# df['Age'].fillna(df['Age'].median(), inplace= True)
```

```python
df['Age']= pd.cut(df['Age'], bins= [0,12,18,35,60,80], labels= ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior'])

age_counts= df.groupby(['Age', 'Survived']).size().unstack(fill_value=0)
age_counts.plot(kind= 'bar', stacked= False)
plt.title('Survival Counts by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Passengers')
plt.legend(['Did not survive(0)', 'Survived(1)'])
plt.show()
```

```python
embarked_counts=df.groupby(['Embarked', 'Survived']).size().unstack(fill_value= 0)
embarked_counts.plot(kind= 'bar', stacked= False)
```

```python
fig, ax= plt.subplots(figsize= (8,6))
width= 0.35
x= np.arange(len(embarked_counts))
ax.bar(x- width/2, height= embarked_counts[0], width= width, label= 'Did not survive(0)')
ax.bar(x+ width/2, height= embarked_counts[1], width= width, label= 'Survived(1)')

ax.set_title('Survival Counts by Embarkation Point')
ax.set_xlabel('Embarkation Point')
ax.set_ylabel('Number of Passengers')
```

```python
print("Key Insights:")
print("- Overall survival rate: {:.2f}%".format(df['Survived'].mean()*100))
print("- Female survival rate: {:.2f}%".format(df[df['Sex']=='female']['Survived'].mean()*100))
print("- Male survival rate: {:.2f}%".format(df[df['Sex']=='male']['Survived'].mean()*100))
print("- 1st Class survival rate: {:.2f}%".format(df[df['Pclass']==1]['Survived'].mean()*100))
print("- 3rd Class survival rate: {:.2f}%".format(df[df['Pclass']==3]['Survived'].mean()*100))
```
